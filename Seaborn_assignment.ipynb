#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(5)
y = np.random.rand(5)
```

```python
## Scatter plot

Use: Display the relationship between two numerical variables.

sns.scatterplot(x,y,color = 'r')
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.show()
```

```python
## line plot

Use: Show the trend or relationship between two numeric variables over a continuous interval.

sns.lineplot(x,y,color = 'r')
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.show()
```

```python
## Barplot

Use: Display the distribution of a categorical variable and compare the values of a numeric variable for each category.

plt.figure(figsize=(10,8))
sns.barplot(x,y,color = 'r')
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.tight_layout()
plt.show()
```

```python
## Box plot
import pandas as pd

data = {
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'Sales', 'Research'],
    'Salary': [50000, 60000, 75000, 55000, 70000, 80000]
}

df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Salary', data=df)
plt.title('Salary Distribution by Department')
plt.show()
```

```python
## heatmap

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

#### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
p = sns.load_dataset("fmri")
p
```

```python
sns.lineplot(x = p.timepoint, y = p.signal)
plt.xlabel("timepoint")
plt.ylabel('signal')
plt.title('Distribution between timepoint and signal')
plt.show()
```

#### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.boxplot(x = titanic['pclass'], y = titanic['age'])
plt.xlabel("pclass")
plt.ylabel('age')
plt.title("Box plot 1 ")
plt.show()
```

```python
sns.boxplot(x = titanic['pclass'], y = titanic['fare'])
plt.xlabel("pclass")
plt.ylabel('Fare')
plt.title("Box plot 2")
plt.show()
```

#### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
sns.histplot(x = diamonds['price'], hue = diamonds['cut'])
plt.xlabel("Price")
plt.title("Diamonds")
plt.show()
```

#### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
